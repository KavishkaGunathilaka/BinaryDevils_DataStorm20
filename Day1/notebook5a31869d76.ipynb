```r
# set.seed(150)  # Setting Seed 
classifier_cl <- naiveBayes(Reservation_Status ~ ., data = hotelData) 
# classifier_cl 
  
# Predicting on test data' 
y_pred <- predict(classifier_cl, newdata = data_test) 
y_pred
# Confusion Matrix 
# cm <- table(data_test$Reservation_Status, y_pred) 
# cm
```

```r
Result = data.frame(data_test$Reservation.id, as.numeric(factor(y_pred, levels = c("Check-In","Canceled","No-Show", ordered=TRUE))))
names(Result) <- c('Reservation-id', 'Reservation_status')
head(Result)
write.csv(Result,"/kaggle/working/out.csv", row.names = FALSE)
nrow(Result)
```
