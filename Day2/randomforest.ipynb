```r
hotelData <- read.csv("../input/data-storm-20/Hotel-A-train.csv")
data_test <- read.csv("../input/data-storm-20/Hotel-A-test.csv")
data_valid <- read.csv("../input/data-storm-20/Hotel-A-validation.csv")
```

```r
data_test$Staying_Days <- as.numeric(as.Date(as.character(data_test$Expected_checkout), format="%m/%d/%Y")-
                  as.Date(as.character(data_test$Expected_checkin), format="%m/%d/%Y"))
hotelData$Staying_Days <- as.numeric(as.Date(as.character(hotelData$Expected_checkout), format="%m/%d/%Y")-
                  as.Date(as.character(hotelData$Expected_checkin), format="%m/%d/%Y"))

hotelData = subset(hotelData, select = -c(Booking_date, Gender, Expected_checkout, Expected_checkin))
data_test = subset(data_test, select = -c(Booking_date, Gender, Expected_checkout, Expected_checkin))
summary(data_test)
```

```r
rf <- randomForest(Reservation_Status ~ .,data=hotelData, ntree=1, importance=TRUE)
```

```r
pred = predict(rf, newdata=data_test)
pred
```

```r
Result = data.frame(data_test$Reservation.id, as.numeric(factor(pred, levels = c("Check-In","Canceled","No-Show", ordered=TRUE))))
names(Result) <- c('Reservation-id', 'Reservation_status')
head(Result)
write.csv(Result,"/kaggle/working/out.csv", row.names = FALSE)
nrow(Result)
```
